# 등록할 신조어 취합

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/3조_XXXX/04. 기타자료/data/'
```

```python
edit_nouns = set()
with open(file_path + 'okt_soy_nouns.txt', 'r') as f:
    edit_nouns.update(f.read().split('\n'))

print(f'등록할 신조어 명사는 {len(edit_nouns)}개 입니다.')
```

```python
edit_etcs = set()
with open(file_path + 'okt_soy_etc.txt', 'r') as f:
    edit_etcs.update(f.read().split('\n'))

print(f'등록할 신조어 명사는 {len(edit_etcs)}개 입니다.')
```

# 사전에 추가하기

## 명사

```python
dic_path = '/content/packages/konlpy/java/tmp1/org/openkoreantext/processor/util/'
with open(dic_path + 'noun/profane.txt', 'a') as f:
    for noun in edit_nouns:
        f.write(noun + '\n')
```

## 형용사

```python
dic_path = '/content/packages/konlpy/java/tmp1/org/openkoreantext/processor/util/'
with open(dic_path + 'adjective/adjective.txt', 'a') as f:
    for adj in edit_etcs:
        f.write(adj + '\n')
```

## 접미사

```python
suffix_lst = ['련']
dic_path = '/content/packages/konlpy/java/tmp1/org/openkoreantext/processor/util/'
with open(dic_path + 'substantives/suffix.txt', 'a') as f:
    for suffix in suffix_lst:
        f.write(suffix + '\n')
```
